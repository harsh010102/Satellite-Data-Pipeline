```python
from geopy.geocoders import Nominatim

def get_lat_long(location_name):

    geolocator = Nominatim(user_agent="location_to_lat_long")

    try:

        location = geolocator.geocode(location_name)

        if location:

            latitude = location.latitude
            longitude = location.longitude
            return latitude, longitude
        else:
            return None
    except Exception as e:
        print(f"Error: {e}")
        return None

if __name__ == "__main__":

    location_names = [
        "New York, USA",
        "Paris, France",
        "Sydney, Australia",
        "Tokyo, Japan",
        "London, UK",
        "Ahmedabad,India"
    ]

    for location_name in location_names:
        coordinates = get_lat_long(location_name)

        if coordinates:
            print(f"Location: {location_name}, Latitude: {coordinates[0]}, Longitude: {coordinates[1]}")
        else:
            print(f"Location not found or an error occurred for: {location_name}")
```

```python
import geopandas as gpd
from shapely.geometry import Point
import json

def create_geojson(lon, lat, radius_km=500):

    point = Point(lon, lat)


    buffer = point.buffer(radius_km / 111.32)

    #GeoDataFrame to store the geometry
    gdf = gpd.GeoDataFrame({'geometry': [buffer]})

    # GeoJSON dictionary
    geojson_dict = json.loads(gdf.to_json())

    return geojson_dict

gj=create_geojson(72.5797068,23.0216238)
print(json.dumps(gj, indent=2))
```

```python

if __name__ == "__main__":
    # Input longitude and latitude (e.g., for a location in Paris)
    input_longitude = 2.3522
    input_latitude = 48.8566

    # Generate GeoJSON dictionary
    geojson = create_geojson(input_longitude, input_latitude)

    # Print the GeoJSON dictionary
    print(json.dumps(geojson, indent=2))
```
